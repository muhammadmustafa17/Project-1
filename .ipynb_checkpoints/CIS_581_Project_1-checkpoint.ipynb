Import Required Libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

Importing dataset

```python
train_data = pd.read_csv("train.dat", sep=" ", header=None, names=["Week", "Cases"])
test_data = pd.read_csv("test.dat", sep=" ", header=None, names=["Week", "Cases"])

print("Train Data:\n", train_data)
print("\nTest Data:\n", test_data)
```

Standard Scaling

```python
scaler_x = StandardScaler()
scaler_y = StandardScaler()

X_train = scaler_x.fit_transform(train_data[['Week']].values)
y_train = scaler_y.fit_transform(train_data[['Cases']].values).flatten()

X_test = scaler_x.transform(test_data[['Week']].values)
y_test = scaler_y.transform(test_data[['Cases']].values).flatten()
```

12-Fold Cross-Validation for Polynomial Degree Selection

```python
kf = KFold(n_splits=12, shuffle=True, random_state=42)

degrees = np.arange(0, 29)
cv_rmse_train = []
cv_rmse_test = []

for d in degrees:
    model = make_pipeline(PolynomialFeatures(d), Ridge(alpha=0))  # No regularization
    train_rmse = -cross_val_score(model, X_train, y_train, scoring="neg_root_mean_squared_error", cv=kf).mean()
    test_rmse = -cross_val_score(model, X_train, y_train, scoring="neg_root_mean_squared_error", cv=kf).mean()
    
    cv_rmse_train.append(train_rmse)
    cv_rmse_test.append(test_rmse)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(degrees, cv_rmse_train, label="Train RMSE")
plt.plot(degrees, cv_rmse_test, label="Test RMSE")
plt.xlabel("Polynomial Degree")
plt.ylabel("RMSE")
plt.legend()
plt.title("12-Fold CV: Polynomial Degree vs RMSE")
plt.show()
```

Select optimal d*

```python
d_star = degrees[np.argmin(cv_rmse_test)]
print(f"Optimal polynomial degree (d*): {d_star}")
```

12-Fold CV for Regularization

```python
lambdas = np.exp(np.arange(-30, 11, 2))  # Exponents from -30 to 10
cv_rmse_lambda = []

for lmb in lambdas:
    model = make_pipeline(PolynomialFeatures(28), Ridge(alpha=lmb))
    test_rmse = -cross_val_score(model, X_train, y_train, scoring="neg_root_mean_squared_error", cv=kf).mean()
    
    cv_rmse_lambda.append(test_rmse)
```

Plot RMSE vs Lambda

```python
plt.figure(figsize=(10, 5))
plt.plot(np.log(lambdas), cv_rmse_lambda, marker='o')
plt.xlabel("log(Lambda)")
plt.ylabel("CV RMSE")
plt.title("12-Fold CV: Regularization Parameter vs RMSE")
plt.show()
```

Select optimal lambda*

```python
lambda_star = lambdas[np.argmin(cv_rmse_lambda)]
print(f"Optimal regularization parameter (lambda*): {lambda_star:.5f}")
```

Train Final Model

```python
poly_d_star = make_pipeline(PolynomialFeatures(d_star), Ridge(alpha=0))
poly_d_star.fit(X_train, y_train)

poly_reg_28 = make_pipeline(PolynomialFeatures(28), Ridge(alpha=lambda_star))
poly_reg_28.fit(X_train, y_train)

y_train_pred_d_star = poly_d_star.predict(X_train)
y_test_pred_d_star = poly_d_star.predict(X_test)

y_train_pred_28 = poly_reg_28.predict(X_train)
y_test_pred_28 = poly_reg_28.predict(X_test)

rmse_train_d_star = mean_squared_error(y_train, y_train_pred_d_star, squared=False)
rmse_test_d_star = mean_squared_error(y_test, y_test_pred_d_star, squared=False)

rmse_train_28 = mean_squared_error(y_train, y_train_pred_28, squared=False)
rmse_test_28 = mean_squared_error(y_test, y_test_pred_28, squared=False)

print(f"Final Model RMSEs:")
print(f" - d*={d_star}:\n Train RMSE = {rmse_train_d_star:.4f},\t Test RMSE = {rmse_test_d_star:.4f}")
print(f" - d=28, lambda*={lambda_star:.5f}:\n Train RMSE = {rmse_train_28:.4f},\t Test RMSE = {rmse_test_28:.4f}")
```

Plot Final Results

```python
weeks_plot = np.linspace(1, 91, 300).reshape(-1, 1)
weeks_scaled = scaler_x.transform(weeks_plot)

cases_pred_d_star = scaler_y.inverse_transform(poly_d_star.predict(weeks_scaled).reshape(-1, 1))
cases_pred_28 = scaler_y.inverse_transform(poly_reg_28.predict(weeks_scaled).reshape(-1, 1))

plt.figure(figsize=(10, 6))
plt.scatter(train_data["Week"], train_data["Cases"], label="Training Data", color="blue", alpha=0.5)
plt.plot(weeks_plot, cases_pred_d_star, label=f"Polynomial (d={d_star})", color="red")
plt.xlabel("Week Number")
plt.ylabel("New COVID-19 Cases")
plt.title(f"Best-Fit Polynomial Curve (d={d_star})")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(train_data["Week"], train_data["Cases"], label="Training Data", color="blue", alpha=0.5)
plt.plot(weeks_plot, cases_pred_28, label=f"Regularized (d=28, ={lambda_star:.5f})", color="green")
plt.xlabel("Week Number")
plt.ylabel("New COVID-19 Cases")
plt.title(f"Regularized Polynomial Curve (d=28, ={lambda_star:.5f})")
plt.legend()
plt.grid(True)
plt.show()
```
