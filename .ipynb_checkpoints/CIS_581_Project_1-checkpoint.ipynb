Import Required Libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import Ridge
from sklearn.model_selection import KFold, cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

Importing dataset

```python
train_data = pd.read_csv("train.dat", sep=" ", header=None, names=["Week", "Cases"])
test_data = pd.read_csv("test.dat", sep=" ", header=None, names=["Week", "Cases"])

print("Training Data:")
print(train_data)

print("\nTest Data:")
print(test_data)
```

Histogram of Test.dat and Train.dat

```python
plt.figure(figsize=(10, 5))
plt.hist(train_data["Cases"], bins=20, color="yellow", alpha=0.6, label="Training Data")
plt.hist(test_data["Cases"], bins=20, color="green", alpha=0.6, label="Test Data")

plt.xlabel("Number of Cases")
plt.ylabel("Frequency")
plt.title("Histogram of COVID-19 Cases in Training and Test Data")
plt.legend()
plt.grid(True)

plt.show()
```

Scaling, Reshaping and Trainging Test and Train data

```python
scaler_X = StandardScaler()
scaler_y = StandardScaler()

train_X = train_data["Week"].values.reshape(-1, 1)
train_y = train_data["Cases"].values.reshape(-1, 1)

test_X = test_data["Week"].values.reshape(-1, 1)
test_y = test_data["Cases"].values.reshape(-1, 1)

train_X_scaled = scaler_X.fit_transform(train_X)
train_y_scaled = scaler_y.fit_transform(train_y)

test_X_scaled = scaler_X.transform(test_X)
test_y_scaled = scaler_y.transform(test_y)
```

Perform 12-Fold Cross-Validation to Find the Best Polynomial Degree (d)*

```python
degrees = range(0, 29)  # 0 to 28

kf = KFold(n_splits=12, shuffle=False)

train_errors = []
test_errors = []

for d in degrees:
    poly = PolynomialFeatures(degree=d)
    
    model = make_pipeline(poly, Ridge(alpha=0))  # No regularization yet

    train_rmse = np.sqrt(-cross_val_score(model, train_X_scaled, train_y_scaled, scoring="neg_mean_squared_error", cv=kf).mean())
    test_rmse = np.sqrt(-cross_val_score(model, train_X_scaled, train_y_scaled, scoring="neg_mean_squared_error", cv=kf).mean())

    train_errors.append(train_rmse)
    test_errors.append(test_rmse)

best_d = degrees[np.argmin(test_errors)]
print(f"Optimal polynomial degree (d*): {best_d}")
```

Train the Model Using the Best Polynomial Degree (d)*

```python
best_poly = PolynomialFeatures(degree=best_d)
model = make_pipeline(best_poly, Ridge(alpha=0))
model.fit(train_X_scaled, train_y_scaled)

train_pred = model.predict(train_X_scaled)
test_pred = model.predict(test_X_scaled)

final_train_rmse = np.sqrt(mean_squared_error(train_y_scaled, train_pred))
final_test_rmse = np.sqrt(mean_squared_error(test_y_scaled, test_pred))

print(f"Final Train RMSE: {final_train_rmse}")
print(f"Final Test RMSE: {final_test_rmse}")
```

Find the Best Regularization Parameter (λ) for d = 28*

```python
lambda_values = np.exp(np.arange(-30, 11, 2))

best_lambda = None
best_rmse = float("inf")

for lam in lambda_values:
    model = make_pipeline(PolynomialFeatures(28), Ridge(alpha=lam))
    test_rmse = np.sqrt(-cross_val_score(model, train_X_scaled, train_y_scaled, scoring="neg_mean_squared_error", cv=kf).mean())

    if test_rmse < best_rmse:
        best_rmse = test_rmse
        best_lambda = lam

print(f"Optimal Regularization Parameter (λ*): {best_lambda}")
```

Train the Final Regularized Model

```python
final_model = make_pipeline(PolynomialFeatures(28), Ridge(alpha=best_lambda))
final_model.fit(train_X_scaled, train_y_scaled)

final_train_pred = final_model.predict(train_X_scaled)
final_test_pred = final_model.predict(test_X_scaled)

final_train_rmse = np.sqrt(mean_squared_error(train_y_scaled, final_train_pred))
final_test_rmse = np.sqrt(mean_squared_error(test_y_scaled, final_test_pred))

print(f"Final Regularized Train RMSE: {final_train_rmse}")
print(f"Final Regularized Test RMSE: {final_test_rmse}")
```

Plot Results

```python
poly = PolynomialFeatures(degree=3)  
train_X_poly = poly.fit_transform(train_X)

scaler_X = StandardScaler()
train_X_scaled = scaler_X.fit_transform(train_X_poly)

scaler_y = StandardScaler()
train_y_scaled = scaler_y.fit_transform(train_y.reshape(-1, 1))

# Fit models
model.fit(train_X_scaled, train_y_scaled)
final_model.fit(train_X_scaled, train_y_scaled)

# Transform new data
X_range = np.linspace(1, 91, 300).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
X_range_scaled = scaler_X.transform(X_range_poly)

# Predict using models
y_pred_d = model.predict(X_range_scaled)
y_pred_lambda = final_model.predict(X_range_scaled)

# Inverse transform predictions
y_pred_d_original = scaler_y.inverse_transform(y_pred_d)
y_pred_lambda_original = scaler_y.inverse_transform(y_pred_lambda)

# ---- First Graph: Best d* ----
plt.figure(figsize=(10, 5))  # Create new figure
plt.scatter(train_X, train_y, color="blue", label="Training Data", alpha=0.5)
plt.plot(X_range, y_pred_d_original, color="red", label=f"Best d*={best_d}")

plt.xlabel("Week")
plt.ylabel("New Cases")
plt.legend()
plt.title("COVID-19 Prediction - Best d* Model")
plt.show()

# ---- Second Graph: Regularized Model ----
plt.figure(figsize=(10, 5))  # Create new figure
plt.scatter(train_X, train_y, color="blue", label="Training Data", alpha=0.5)
plt.plot(X_range, y_pred_lambda_original, color="green", label=f"Regularized d=28, λ*={best_lambda}")

plt.xlabel("Week")
plt.ylabel("New Cases")
plt.legend()
plt.title("COVID-19 Prediction - Regularized Model")
plt.show()
```
